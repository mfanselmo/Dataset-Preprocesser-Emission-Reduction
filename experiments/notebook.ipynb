```python
from classifiers.time_series.fcn import Classifier_FCN
from classifiers.time_series.resnet import Classifier_RESNET
from classifiers.time_series.mlp import Classifier_MLP
from classifiers.images.cnn_small import Classifier_CNN_Small
from classifiers.images.cnn_big import Classifier_CNN_Big



from utils.run_utils import  save_results, format_result


import json
import tqdm



models = {
    'FCN': Classifier_FCN,
    'RESNET': Classifier_RESNET,
    'MLP': Classifier_MLP,
    'CNN_SMALL': Classifier_CNN_Small,
    'CNN_BIG': Classifier_CNN_Big,
}
```

```python
def skip_options(options):
    if options['iteration'] >= 5:
        return True


    return False

def run_experiment(options, dataset_type):
    model = models[options['model']](options, verbose = False)
    model.fit()

    result = model.predict()

    format_result(options, result)
    save_results(options, result, dataset_type)

```

```python
dataset_type = 'time_series'
experiments = ['e00']


for experiment in experiments:
    path = f"settings/{dataset_type}/{experiment}_options.json"
    combinations = []
    with open(path, "r") as f:
        combinations = json.loads(f.read())


    for i in range(5):
        for options in tqdm.tqdm(combinations):
            if skip_options(options): continue
            print(options)

            run_experiment(options, dataset_type)

            options['iteration'] += 1
            json_object = json.dumps(combinations, indent = 4)
            with open(path, "w") as outfile:
                outfile.write(json_object)

```
